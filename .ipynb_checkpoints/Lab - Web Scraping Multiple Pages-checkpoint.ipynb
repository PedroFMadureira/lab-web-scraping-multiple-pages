```python
import requests
from bs4 import BeautifulSoup
from time import sleep
from random import randint
import pandas as pd
```

```python
url = 'https://radiocomercial.pt/programas/tnt-todos-no-top'
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text,'html.parser')
print(soup)
```

```python
songs_ref = '#page-content-wrapper > div > div:nth-child(2) > div > div > div:nth-child(1) > div.wrapper.rounded-site.tm-audio-data > div.cover.rounded-site-top.tnt-box-cover > div > div.songTitle'
```

```python
artist_ref = '#page-content-wrapper > div > div:nth-child(2) > div > div > div:nth-child(1) > div.wrapper.rounded-site.tm-audio-data > div.cover.rounded-site-top.tnt-box-cover > div > div.songArtist'
```

```python
soup.select(songs_ref)[0].text.strip()
```

```python
soup.select(artist_ref)[0].text.strip()
```

```python
i = 1
songs = []
artists = []

while i <= 20:
    songs_ref = '#page-content-wrapper > div > div:nth-child(2) > div > div > div:nth-child(' + str(i) + ') > div.wrapper.rounded-site.tm-audio-data > div.cover.rounded-site-top.tnt-box-cover > div > div.songTitle'
    song = soup.select(songs_ref)[0].text.strip()
    songs.append(song)
    artist_ref = '#page-content-wrapper > div > div:nth-child(2) > div > div > div:nth-child(' + str(i) + ') > div.wrapper.rounded-site.tm-audio-data > div.cover.rounded-site-top.tnt-box-cover > div > div.songArtist'
    artist = soup.select(artist_ref)[0].text.strip()
    artists.append(artist)
    print(i, song, '-', artist)
    i += 1
```

```python
songs = pd.DataFrame(songs)
artists = pd.DataFrame(artists)

Top20_comercial = pd.concat([songs,artists],axis=1)
Top20_comercial.columns = ['song','artist']
```
